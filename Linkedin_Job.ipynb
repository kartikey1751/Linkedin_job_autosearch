```python
import numpy as np
import pandas as pd 
import os
import warnings
warnings.filterwarnings("ignore")
import requests
from bs4 import BeautifulSoup
```

```python
url_Jobs = "https://www.linkedin.com/jobs-guest/jobs/api/seeMoreJobPostings/search?keywords=&location=Worldwide&locationId=&geoId=92000000&f_TPR=&f_JT=F%2CC&f_WT=2&start=25"
```

```python
locations = []
titles = []
countries = []    
links = []
company_name = []


for i in range(0,150,1):
    url = url_Jobs+"&start="+str(i)
    request = requests.get(url)
    data = request.text
    soup = BeautifulSoup(data)

    # Extracting job listings
    job_listings = soup.find_all("li")
    for job in job_listings:
        # Extracting location
        location_tag = job.find("span", class_="job-search-card__location")
        location = location_tag.text.strip() if location_tag else "NA"
        locations.append(location)

        # Extracting title
        title_tag = job.find("h3", class_="base-search-card__title")
        title = title_tag.text.strip() if title_tag else "NA"
        titles.append(title)

        # Extracting country
        country_tag = job.find("span", class_="job-search-card__location")
        country = country_tag.text.strip().split(',')[-1].strip() if country_tag else "NA"
        countries.append(country)

        # Extracting job link
        link_tag = job.find("a", class_="base-card__full-link")
        link = link_tag["href"] if link_tag else "NA"
        links.append(link)

        # Extracting company name
        company_name_tag = job.find("a", class_="hidden-nested-link")
        company = company_name_tag.text.strip() if company_name_tag else "NA"
        company_name.append(company)
```

```python
print(len(locations))
print(len(titles))
print(len(countries))
print(len(links))
print(len(company_name))
```

```python
df1 = pd.DataFrame(titles,columns=["Title"])
df1["Country"] = countries
df1["Links"] = links
df1["Company_Name"] = company_name

df1
```
