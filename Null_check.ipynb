```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

import os
import warnings
warnings.filterwarnings("ignore")
import requests
from bs4 import BeautifulSoup
```

```python
url_Jobs = "https://www.linkedin.com/jobs-guest/jobs/api/seeMoreJobPostings/search?keywords=&location=Worldwide&locationId=&geoId=92000000&f_TPR=&f_JT=F%2CC&f_WT=2"
```

```python
locations = []
titles = []
countries = []    
links = []
company_name = []


for i in range(0,3,1):
    url = url_Jobs+"&start="+str(i)
    request = requests.get(url)
    data = request.text
    soup = BeautifulSoup(data)
    location = soup.findAll("span", {"class":"job-search-card__location"})
    for i in location:
        if i is not None:
            locations.append(i.text.replace("\n"," ").strip(" "))   
        else:
            locations.append("NA")
    title = soup.findAll("h3",{"class":"base-search-card__title"})
    for i in title:
        if i is not None:
            titles.append(i.text.replace("\n"," ").strip(" ")) 
        else:
            titles.append("NA")
    country = soup.findAll("span", {"class":"job-search-card__location"})
    for i in country:
        if i is not None:
            countries.append(i.text.replace("\n"," ").strip(" ").split(',')[-1])
        else:
            countries.append("NA")
    job_links = soup.findAll("a",{"class":"base-card__full-link"})
    for i in job_links:
        if i is not None:
            links.append(i["href"])
        else:
            links.append("NA")
    names = soup.findAll("a",{"class":"hidden-nested-link"})
    for i in names:
        if i is not None:
            company_name.append(i.text.replace("\n"," ").strip(" ")) 
        else:
            company_name.append("NA")
```

```python
print(len(locations))
print(len(titles))
print(len(countries))
print(len(links))
print(len(company_name))
```

```python
df1 = pd.DataFrame(titles,columns=["Title"])
df1["Location"] = locations
df1["Country"] = countries
df1["Links"] = links
df1["Company_Name"] = company_name

df1
```
